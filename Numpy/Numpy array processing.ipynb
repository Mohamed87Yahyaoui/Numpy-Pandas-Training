```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.arange(-4,4,0.2)
points
```

```python
dx , dy = np.meshgrid(points,points)
```

```python
z = np.sin(dx) + np.sin(dy)
z
```

```python
plt.plot(z)
```

```python
plt.imshow(z)
plt.colorbar()
plt.title("sin(dx)+sin(dy)")
```
